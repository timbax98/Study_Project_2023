# **Study Project:** *Transformer model for prediction of grasping movements*

## Import packages

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
```

## Prediction

### Import trained model

```python
transformer = load_model("transformer")
```

### Generate predictions

```python
# Generate dummy novel data (one batch) with input shape from training data
batches = 2
batch_size = 3
sequence_length = 126
input_dim = 8
input_data = tf.random.uniform((batch_size, sequence_length, input_dim))

# Obtain predictions
predictions = transformer.predict(input_data)
print(f"Shape: {predictions.shape}, \nPredictions: \n{predictions}")
```
