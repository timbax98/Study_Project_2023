```python
# Grab frame data
frames = data["frames"]
bounding_boxes = []
```

```python
# Grab frames
for frame in frames:
    figures = frame['figures']
    for figure in figures:
        points = figure['geometry']['points']['exterior']
        x1, y1 = points[0]
        x2, y2 = points[1]
        bounding_box = (x1, y1, x2, y2)
        bounding_boxes.append(bounding_box)
```

```python
# Print bounding box positions
for i, box in enumerate(bounding_boxes):
    print(f"Frame {i}: Bounding box: {box}")
```

```python
def calculate_middle_point(bbox):
    x1, y1, x2, y2 = bbox
    middle_x = (x1 + x2) / 2
    middle_y = (y1 + y2) / 2
    return middle_x, middle_y



for frame_idx, bbox in enumerate(bounding_boxes):
    print(f"Frame {frame_idx + 1}:")
    middle_x, middle_y = calculate_middle_point(bbox)
    print(f"Middle point: ({middle_x}, {middle_y})")
```
